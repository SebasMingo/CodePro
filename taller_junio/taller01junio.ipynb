```python
# Creando una lista enlazada simple

class Nodo:
    '''El nodo siempre consta de dos partes carga/dato 
    enlace/apuntador'''
    def __init__(self, valor):
        self.valor = valor # Dato / carga
        self.siguiente = None # Enlace / apuntador

class ListaEnlazada:
    '''Todas las listas enlazadas inician con la 
    cabeza y siempre apuntan al nulo'''
    def __init__(self):
        self.cabeza = None

    def insertar_nodo_al_inicio(self, valor):
        nuevo_nodo = Nodo(valor) # Se crea un nodo usando la clase Nodo
        nuevo_nodo.siguiente = self.cabeza # Se inserta el primer nodo 
        # y nuevo_nodo.siguiente = None (self.cabeza)
        self.cabeza = nuevo_nodo # La cabeza hacia el nuevo nodo

    def insertar_segun_posicion(self, valor, posicion): #para que se inserte donde quiero(posicion)
        nuevo_nodo = Nodo(valor)#creamos un nodo con la clase Nodo
        #Verificamos

        if posicion == 0: #si es la cabeza 
            nuevo_nodo.siguiente = self.cabeza
            self.cabeza = nuevo_nodo
        else:
            nodo_anterior = self.cabeza
            for __ in range(posicion - 1):
                if nodo_anterior is None:
                    raise IndexError('posicion fuera de rango')
                nodo_anterior = nodo_anterior.siguiente

            nuevo_nodo.siguiente = nodo_anterior.siguiente
            nodo_anterior.siguiente = nuevo_nodo
    
    def insertar_al_final(self, valor):
        nuevo_nodo = Nodo(valor)

        if self.cabeza is None: #en caso de que nuestra cabeza sea nuestro unico elemento el nuevo nodo se agrega al final.
            self.cabeza = nuevo_nodo
            return
        
        nodo_actual = self.cabeza
        #recorremos hasta nuestro ultimo nodo
        while nodo_actual.siguiente:
            nodo_actual = nodo_actual.siguiente

        nodo_actual.siguiente = nuevo_nodo #insertamos nuestro nuevo al final de la lista enlazada

    #ELIMINACION DE NODO
    def eliminar_al_inicio(self):
        '''Eliminia el primer nodo'''
        if self.cabeza is not None:
            self.cabeza = self.cabeza.siguiente
    
    def eliminiar_segun_posicion(self, posicion):
        if self.cabeza is None:
            return
        
        if posicion == 0:
            self.cabeza = self.cabeza.siguiente
            return
        

        nodo_anterior = self.cabeza
        for _ in range(posicion - 1):
            if nodo_anterior.siguiente is None:
                return
            
            nodo_anterior = nodo_anterior.siguiente
        
        if nodo_anterior.siguiente is None:
            return
        
        nodo_anterior.siguiente = nodo_anterior.siguiente.siguiente

    def eliminar_al_final(self):
        '''apuntadores'''

        #si la lista esta vacia no hay nada que hacer
        if self.cabeza is None:
            return
        
        if self.cabeza.siguiente is None:
            self.cabeza = None
            return
        
        #recorremos nuestra lista enlazada
        nodo_actual = self.cabeza
        #recorremos hasta el penultimo nodo
        while nodo_actual.siguiente.siguiente:
            nodo_actual = nodo_actual.siguiente

            #establecemos el nodo actual.siguiente en None
            nodo_actual.siguiente = None


    def imprimir_lista_enlazada(self):
        nodo_actual = self.cabeza # Digo que mi nodo actual es la cabeza

        while nodo_actual: # Recorremos mientras exista un nodo actual
            print(nodo_actual.valor, end="->") # Imprimo el nodo y muestro el valor
            nodo_actual = nodo_actual.siguiente # Recorriendo los nodos
        print("None") # Deja de existir nodos y al final muestro el none o nulo

lista_enlazada = ListaEnlazada()
print(lista_enlazada.cabeza)
lista_enlazada.insertar_nodo_al_inicio(5)
lista_enlazada.insertar_nodo_al_inicio(10)
lista_enlazada.insertar_nodo_al_inicio(90)

lista_enlazada.insertar_segun_posicion(32, 1)

lista_enlazada.insertar_al_final(37)
lista_enlazada.insertar_al_final(40)

lista_enlazada.imprimir_lista_enlazada()
lista_enlazada.eliminar_al_inicio()
lista_enlazada.imprimir_lista_enlazada()
lista_enlazada.eliminiar_segun_posicion(3)
lista_enlazada.imprimir_lista_enlazada()
```

```python
# Listas doblemente enlazadas

class NodoDoble:
    def __init__(self, valor):
        self.valor = valor
        self.siguiente = None
        self.anterior = None
        
class ListaDoblementeEnlazada:
    def __init__(self):
        self.cabeza = None
        
# Insertar nodo al inicio
    def insertar_nodo_al_inicio(self, valor):
        nuevo_nodo = NodoDoble(valor)
    
    # Si la lista esta vacia el nuevo nodo se convierte en la cabeza
        if self.cabeza is None:
            self.cabeza = nuevo_nodo
            return
        
        else:
            # Si la lista no esta vacia, el nuevo nodo se enlaza con la cabeza actual
            nuevo_nodo.siguiente = self.cabeza
            self.cabeza.anterior = nuevo_nodo
            self.cabeza = nuevo_nodo # El nuevo nodo se convierte en la nueva cabeza
            
    def imprimir_adelante(self):
        '''Imprime la lista enlazada en orden, desde la cabeza hasta el final'''
        nodo_actual = self.cabeza
        while nodo_actual:
            print(nodo_actual.valor, end='<->')
            nodo_actual = nodo_actual.siguiente
        print('None')
        
    def imprimir_atras(self):
        '''Imprime la lista enlazada en orden inverso, desde el final hasta la cabeza'''
        nodo_actual = self.cabeza
        # nos podemos mover hasta el ultimo nodo
        while nodo_actual.siguiente:
            nodo_actual = nodo_actual.siguiente
        
        #desde el final hacia adelante otra vez para imprimir los valores
        while nodo_actual:
            print(nodo_actual.valor, end='<->')
            nodo_actual = nodo_actual.anterior
        print('None')
        
        

lista_doble = ListaDoblementeEnlazada()
lista_doble.insertar_nodo_al_inicio(1)  
lista_doble.insertar_nodo_al_inicio(2)
lista_doble.insertar_nodo_al_inicio(3)  


lista_doble.imprimir_adelante()
lista_doble.imprimir_atras()
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Stacked_linked_list:
    def __init__(self):
        self.top = None

    def push(self, value):
        '''Inserta un nodo en la cima de la pila'''
        new_node = Node(value) #Creamos un nodo con la clase Node
        new_node.next = self.top # 
        self.top = new_node # Actualiza la cima de la pila

    def pop(self):
        if self.is_empty():
            raise IndexError("Error la pila esta vacia")
        nodo_a_eliminar = self.top
        self.top = self.top.next
        return nodo_a_eliminar.value
    
    def peek (self):
        if self.os_empty ():
            raise IndexError("La pila esta vacia")
        return self.top.value
    def is_empty(self):
        return self.top is None
    

pila = Stacked_linked_list()
pila.push(3)
pila.push(23)
pila.push(13)

print('Pila:')
while not pila.is_empty():
    print(pila.pop())
```

```python
#COLAS
class Node:
    def init(self, valor):
        self.valor = valor 
        self.siguiente = None

class Cola_listas_enlazadas:

    def init(self):
        self.cabeza = None
        self.cola = None 

    def enqueue(self, valor):
        nuevo_nodo = Nodo (valor)

        if self.is_empty(): #verificando si la cola esta vacia 
            self.cabeza = nuevo_nodo
        else: #si no esta vacia el siguiente nodo del ultimo nodo apunta al nuevo nodo
            self.cola.siguiente = nuevo_nodo
#el nuevo nodo sea el ultimo nodo
        self.cola = nuevo_nodo

    def dequeue(self):
        if self.is_empty():
            raise IndexError ("error la cola esta vacia")

        nodo_eliminando = self.cabeza
        self.cabeza = self.cabeza.siguiente

        if self.cabeza is None: 
            self.cola = None

        return nodo_eliminando.valor

    def is_empty(self):
        return self.cabeza is None #retorna true si la cabeza de la cola es none (la cola esta vacia)

cola = Cola_listas_enlazadas()
cola.enqueue(2)
cola.enqueue(24)
cola.enqueue(25)
print("cola:")

while not cola.is_empty():
    print(cola.dequeue)
```

```python
from collections import deque

class NodoArbol:
    def __init__ (self, valor):
        self.valor = valor 
        self.izquierda = None
        self.derecha = None

class ArbolBinario:
    def __init__ (self, valor_raiz):
        self.raiz = NodoArbol(valor_raiz)

    def insertar_izquierda(self, nodo_padre, valor):
        #insertamos un nodo a la izquierda del nodo padre
        nuevo_nodo = NodoArbol(valor)
        nodo_padre.izquierda = nuevo_nodo
    
    def insertar_derecha(self, nodo_padre, valor):
        #insertamos un nodo a la derecha del nodo padre
        nuevo_nodo = NodoArbol(valor)
        nodo_padre.derecha = nuevo_nodo
    
    def imprimir_arbol(self): 
        #Breadth-First Search
        #recorrer por niveles el arbol
        if not self.raiz:
            return
        cola = deque([self.raiz]) #cola que inicia con la raiz del arbol.
        while cola:
            tamanho_nivel = len(cola) #calcula el tamanho del nivel actual para iterar en los nodos
            for _ in range(tamanho_nivel):
                nodo = cola.popleft()
                print(nodo.valor,end="")
                if nodo.izquierda:
                    cola.append(nodo.izquierda)
                if nodo.derecha:
                    cola.append(nodo.derecha)



#Creamos un arbol
        arbol = ArbolBinario(5)
        arbol.insertar_derecha(arbol.raiz, 6)
        arbol.insertar_izquierda(arbol.raiz, 1)
        arbol.insertar_izquierda(arbol.raiz.izquierda, 7)
        arbol.insertar_derecha(arbol.raiz.izquierda , 9)
        arbol.imprimir_arbol()
```
